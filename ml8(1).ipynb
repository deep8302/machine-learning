```python
import pandas as pd

import matplotlib.pyplot as plt

df=pd.read_csv("salary.csv")
```

```python
#exp
exp=df.iloc[:,0:1].values
```

```python
#sal

sal=df.iloc[:,1:].values
```

```python
# call linesr model just for compare
from sklearn.linear_model import LinearRegression
regr1=LinearRegression()
```

```python
# to transform data we can call poly transform method

from sklearn.preprocessing import PolynomialFeatures
```

```python
#experience transform
poly=PolynomialFeatures(degree=2)
```

```python
#exp fit

newexp=poly.fit_transform(exp)
```

```python
# apply linear regression

trained1=regr1.fit(exp,sal)
```

```python
# now creating linear model again

plt.xlabel("exp")
plt.ylabel("sal")
plt.scatter(exp,sal,label="actual sal")
plt.plot(exp,trained1.predict(exp),label="linear model predicted salary")
plt.legend()
plt.show()
```

```python
# apply new exp data in model

regr2=LinearRegression()
```

```python
#  new exp data apply

trained2=regr2.fit(newexp,sal)
```

```python
# now creating linear model again

plt.xlabel("exp")
plt.ylabel("sal")
plt.scatter(exp,sal,label="actual sal")
plt.plot(sal,trained2.predict(newexp),label="new model")
plt.legend()
plt.show()
```
